Importing the necessary library

```python
import numpy as np
from matplotlib import pyplot as plt
```

#Trajectory of a projectile motion

$y=tanθ⋅x−\frac{gx^2}{2u^2⋅cos^2θ⋅}$

```python
a = 45 #(Angle) degree
u = 30 #(initial velocity) m/s
g = 9.8 #m/s^2
```

```python
#Horizontal range
R = u**2*np.sin(2*a)/g
```

```python
x = np.arange(0,R,0.1)
```

```python
y = x*np.tan(a) - ((9.8)*x*x)/(2*u*u*np.cos(a)*np.cos(a))
```

```python
plt.style.use('fivethirtyeight')
plt.plot(x,y)

plt.title('Trajectory of a projectile')
plt.xlabel('Horizontal Distance')
plt.ylabel('Vertical Distance')

plt.show()
```
