Importing the necessary library

```python
import numpy as np
from matplotlib import pyplot as plt
```

Setting the domain

```python
x = np.arange(0,4*np.pi, np.pi/1800)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
```

Graph Sinx and Cosx

```python
plt.style.use('fivethirtyeight')
plt.plot(x,y, color = 'k', linestyle = '--', linewidth = 3, marker = '.', label = 'Sinx')
plt.plot(x,z, color = '#444444', linewidth = 3, label = 'Cosx')

plt.title('Graph of Sinx and Cos x')
plt.xlabel('x')
plt.ylabel('Sinx and Cosx')

plt.legend()

plt.show()
```
